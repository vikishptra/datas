```python
from pulp import *


x13 = LpVariable('x13', lowBound=0, cat='Integer')
x14 = LpVariable('x14', lowBound=0, cat='Integer')
x15 = LpVariable('x15', lowBound=0, cat='Integer')
x23 = LpVariable('x23', lowBound=0, cat='Integer')
x24 = LpVariable('x24', lowBound=0, cat='Integer')
x25 = LpVariable('x25', lowBound=0, cat='Integer')
x36 = LpVariable('x36', lowBound=0, cat='Integer')
x37 = LpVariable('x37', lowBound=0, cat='Integer')
x38 = LpVariable('x38', lowBound=0, cat='Integer')
x46 = LpVariable('x46', lowBound=0, cat='Integer')
x47 = LpVariable('x47', lowBound=0, cat='Integer')
x48 = LpVariable('x48', lowBound=0, cat='Integer')
x56 = LpVariable('x56', lowBound=0, cat='Integer')
x57 = LpVariable('x57', lowBound=0, cat='Integer')
x58 = LpVariable('x58', lowBound=0, cat='Integer')





#problem inisialisasi
problem = LpProblem('Transshipment problems Retno', LpMinimize)



problem += 16 * x13 + 10 * x14 + 12 * x15 + 15 * x23 + 14 * x24 + 17 * x25 + 6 * x36 + 8 * x37 + 10 * x38 + 7 * x46 + 11 * x47 + 11 * x48 + 4 * x56 + 5 * x57 + 12 * x58





problem += x13 + x14 + x15 == 300
problem += x23 + x24 + x25 == 300
problem += x36 + x46 + x56 == 200
problem += x37 + x47 + x57 == 100
problem += x38 + x48 + x58 == 300
problem += x15 == 300
problem += x13 + x23 == x36 + x37 + x38
problem += x14 + x24 == x46 + x47 + x48 
problem += x15 + x25 == x56 + x57 + x58

problem.solve()

print(LpStatus[problem.status])

print(value(problem.objective))

for s in problem.variables():
    print(s.name, '=', s.varValue)
    
```
